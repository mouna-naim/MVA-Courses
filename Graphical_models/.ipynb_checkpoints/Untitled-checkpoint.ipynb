```python
#Step1: Dimension reduction of the input data
def pca_dimension_reduction(X, num_components):
    pca = PCA(n_components=num_components)
    X_reduced = pca.fit_transform(X)

    # Affichage des composantes principales
    print("Principal Components (Charges):")
    components_df = pd.DataFrame(
        pca.components_,
        columns=feature_names if feature_names else [f"Feature {i}" for i in range(X.shape[1])]
    )
    print(components_df)
    
    return X_reduced, pca

# Step 2: Clustering on X and H recuperation
def kmeans_clustering(X, num_clusters, random_state=42):
    kmeans = KMeans(n_clusters=num_clusters, random_state=random_state, n_init=10)
    kmeans.fit(X)
    labels = kmeans.labels_
    
    # Construction de la matrice H
    n_samples = X.shape[0]
    H = np.zeros((n_samples, num_clusters))
    
    for i in range(n_samples):
        H[i, labels[i]] = 1  # Assigne 1 à la colonne correspondant au cluster de l'échantillon i
    
    return kmeans, H

# Usng R1 initialization 
def kmeans_clustering_with_r1_initialization(X, num_clusters, random_state=42):
    # Initialisation aléatoire R1: attribution aléatoire de points de données à des clusters
    n_samples = X.shape[0]
    labels_r1 = np.random.randint(0, num_clusters, size=n_samples)

    # Calcul des centroïdes initiaux basés sur l'initialisation R1
    initial_centroids = np.array([X[labels_r1 == k].mean(axis=0) for k in range(num_clusters)])
    
    # Filtrer les centroïdes pour s'assurer qu'ils ne contiennent pas de valeurs NaN
    initial_centroids = np.nan_to_num(initial_centroids)

    # Appliquer K-means avec ces centroïdes initiaux
    kmeans = KMeans(n_clusters=num_clusters, init=initial_centroids, n_init=1, random_state=random_state)
    kmeans.fit(X)
    labels = kmeans.labels_
    
    # Construction de la matrice H
    H = np.zeros((n_samples, num_clusters))
    for i in range(n_samples):
        H[i, labels[i]] = 1  # Assigne 1 à la colonne correspondant au cluster de l'échantillon i
    
    return kmeans, H
# Step3: Function to compute centroids
def compute_centroids(X, H):
    num_clusters = H.shape[1] #H est de dimension n x K
    centroids = np.zeros((num_clusters, X.shape[1]))

    for j in range(num_clusters):
        # Sélectionne les points du cluster j
        indices = H[:, j] == 1
        cluster_points = X[indices]
        
        # Calcul du centroïde pour le cluster j
        if len(cluster_points) > 0:
            centroids[j] = cluster_points.mean(axis=0)
    
    return centroids

# Step 4: Clustering on X
def kmeans_with_initial_centroids(X, initial_centroids):
    kmeans = KMeans(n_clusters=len(initial_centroids), init=initial_centroids, n_init=1)
    kmeans.fit(X)
    return kmeans
```

```python
#Normalize data
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Step 1: Dimension reduction of X
num_clusters = 3 # equals to number of clusters
X_reduced, pca_model = pca_dimension_reduction(X_normalized, num_components = num_clusters)

# Step 2: Clustering on X and H recuperation
#kmeans_reduced, H = kmeans_clustering(X_reduced, num_clusters, random_state=42)
kmeans_reduced, H = kmeans_clustering_with_r1_initialization(X, num_clusters, random_state=42)

#Step3: Calculation of centroids: 
Initial_centroids = compute_centroids(X_normalized, H)

#Step 4: Clustering on X using initial centroids
kmeans_final = kmeans_with_initial_centroids(X_normalized, Initial_centroids)

# Affichage des résultats
print("Centroïdes finaux:", kmeans_final.cluster_centers_)
print("Labels finaux:", kmeans_final.labels_)

print("Distorsion (PCA-guided K-means):", kmeans_final.inertia_)
```

```python
# Fonction pour exécuter un clustering K-means classique
def kmeans_classic(X, num_clusters, random_state=42):
    # Normalisation des données
    scaler = StandardScaler()
    X_normalized = scaler.fit_transform(X)

    # Initialisation et ajustement de K-means
    kmeans = KMeans(n_clusters=num_clusters, random_state=random_state, n_init=10)
    kmeans.fit(X_normalized)

    # Retourne le modèle K-means entraîné et les labels
    return kmeans, kmeans.labels_

# Exemple d'appel de la fonction
num_clusters = 3  # Ajuster le nombre de clusters selon les besoins
kmeans_model, labels = kmeans_classic(X, num_clusters)

# Affichage des résultats
print("Centroïdes finaux :", kmeans_model.cluster_centers_)
print("Labels des clusters :", labels)
print("Distorsion :", kmeans_model.inertia_)
```
